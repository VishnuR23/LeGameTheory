```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable = '2023-24')
games = gamefinder.get_data_frames()[0]
```

```python
team_games = games[games['TEAM_ID'].notna()]
```

```python
nba_teams = [
    "ATL", "BOS", "BRK", "CHI", "CHO", "CLE", "DAL", "DEN", "DET", "GSW",
    "HOU", "IND", "LAC", "LAL", "MEM", "MIA", "MIL", "MIN", "NOP", "NYK",
    "OKC", "ORL", "PHI", "PHO", "POR", "SAC", "SAS", "TOR", "UTA", "WAS"
]
```

```python
team_games = team_games[team_games["TEAM_ABBREVIATION"].isin(nba_teams)]
```

```python
keeps = ['GAME_ID', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'PTS', 'FG3A', 'FGA', 'AST', 'REB', 'AST', 'PLUS_MINUS', 'PACE', 'GAME_DATE']
```

```python
team_games.columns
```

```python
team_games['3PA_Rate'] = team_games['FG3A'] / team_games['FGA']
team_games['AST_Rate'] = team_games['AST'] / team_games['PTS']
team_games['PACE_EST'] = team_games['FGA'] + 0.44*team_games['FTA'] - team_games['OREB'] + team_games['TOV']
```
